### Q3-1. 이마의 폭(forehead_width_cm)을 높이(forehead_height_cm)로 나눈 비율(forehead_ratio)에 대해서 평균으로부터 3 표준편차 밖의 경우를 이상치로 정의할 때, 이상치에 해당하는 데이터는 몇 개인가? (답안 예시) 10

```python
df['forehead_ratio'] = df['forehead_width_cm'] / df['forehead_height_cm']
```

```python
mu = df.forehead_ratio.mean()
std = df.forehead_ratio.std()
```

```python
UB = mu + (3*std)
LB = mu - (3*std)
```

```python
sum((df['forehead_ratio'] > UB) | (df['forehead_ratio'] < LB))
```

```python
((df['forehead_ratio'] > UB) | (df['forehead_ratio'] < LB)).sum()
```

```python
# 정답: 3
```
